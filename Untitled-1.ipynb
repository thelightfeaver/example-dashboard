```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_supermarket = pd.read_csv('supermarket_sales.csv')
df_supermarket.head()
```

```python
df_supermarket.info()
```

```python
df_supermarket.isnull().sum()
```

```python
df_supermarket.dtypes
```

```python
df_supermarket.drop(['Invoice ID'], axis=1, inplace=True)
```

```python
df_supermarket.columns = df_supermarket.columns.str.lower()

df_supermarket.columns
```

```python
df_supermarket['date'] = pd.to_datetime(df_supermarket['date'])
df_supermarket['branch'] = df_supermarket['branch'].astype('category')
df_supermarket['city'] = df_supermarket['city'].astype('string')
df_supermarket['customer type'] = df_supermarket['customer type'].astype('category')
df_supermarket['gender'] = df_supermarket['gender'].astype('string')
df_supermarket['product line'] = df_supermarket['product line'].astype('category')
df_supermarket['payment'] = df_supermarket['payment'].astype('category')
df_supermarket.dtypes
```

```python
columns = ["cogs",	"gross margin percentage",	"gross income", "total"]
for col in columns:
    df_supermarket[col] = np.around(df_supermarket[col], 2)

df_supermarket[columns].head()
```

```python
df_supermarket_by_product_line = df_supermarket['product line'].value_counts()
df_supermarket_by_product_line

plt.figure(figsize=(15,  10))
sns.countplot(data=df_supermarket, x='product line')
plt.title('Product Line Count')
plt.xlabel('Product Line')
plt.show()
```

```python
df_method_payment = df_supermarket['payment'].value_counts()
df_method_payment

sns.countplot(data=df_supermarket, x='payment')
plt.title('Payment Method Count')
plt.xlabel('Payment Method')
plt.show()
```

```python
df_product_sum = df_supermarket.pivot_table(index='product line', values=["quantity", 'total'], aggfunc='mean')

sns.barplot(data=df_product_sum, x=df_product_sum.index, y='total')
plt.title('Total Sales by Product Line Mean')
plt.xlabel('Product Line')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()

```

```python
df_data = df_supermarket.pivot_table(index='product line', columns='city', values='quantity', aggfunc='sum') 

sns.heatmap(df_data, annot=True, fmt='d')
plt.title('Quantity by Product Line and City')
plt.xlabel('City')
plt.ylabel('Product Line')
plt.show()
```

```python
df_sellers_by_month = df_supermarket.groupby(df_supermarket['date'].dt.strftime('%B'))['total'].sum().reset_index()
df_sellers_by_month

sns.barplot(data=df_sellers_by_month, x='date', y='total')
plt.title('Total Sales by Month')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()


```

```python
# export df
df_supermarket.to_csv('supermarket_sales_clean.csv', index=False)
```
